# The Model-View-Controller Pattern
### This pattern commonly used in applications separates the responsibilities of the entire application into 3 components

## The Model
### The Model represents the business logic of the application. It manages the data, which can be stored in the Model itself or in a database that only the Model component has access to.

## The View
### The View is what is shown to the user. It can be website page, a mobile application scree or even a simple textual output.

## The Controller
### The Controller accepts user's inputs delegates data handling to a Model component and data representation to a View component
